```python
import sympy as sy
import numpy as np
from sympy.functions import sin,cos, exp, log
import matplotlib.pyplot as plt
```

# Aufgabe 1

Funktion: $\sqrt[3]{2x+ 2} \text{ mit } x \in [-1, 50],\quad x_0 = 3$

```python
# Define the variable and the function to approximate
x = sy.Symbol('x')
f = (2*x+2)**(1/3)

# Factorial function
def factorial(n):
    if n <= 0:
        return 1
    else:
        return n*factorial(n-1)

# Taylor approximation at x0 of the function 'function'
def taylor(function,x0,n):
    i = 0
    p = 0
    while i <= n:
        p = p + (function.diff(x,i).subs(x,x0))/(factorial(i))*(x-x0)**i
        i += 1
    return p
plt.figure(figsize=(9,6))
# Plot results
def plot():
    x_lims = [-1,50]
    x1 = np.linspace(x_lims[0],x_lims[1],800)
    y1 = []
    # Approximate up until 10 starting from 1 and using steps of 2
    for j in range(0,4,1):
        func = taylor(f,3,j)
        print('Taylor expansion at n='+str(j),func)
        for k in x1:
            y1.append(func.subs(x,k))
        plt.plot(x1,y1,label=str(j)+'. Ordnung')
        y1 = []
    # Plot the function to approximate (sine, in this case)
    plt.plot(x1,(2*x1+2)**(1/3),label='f(x)')
    plt.xlim(x_lims)
    plt.ylim([-10,10])
    plt.xlabel('x')
    plt.ylabel('y')
    plt.legend()
    plt.grid(True)
    plt.savefig("Aufgabe_1.pdf")
    plt.show()

plot()
```

Funktion: $x \ln(x) \text{ mit } x \in [0, 10],\quad x_0 = 1$

```python
# Define the variable and the function to approximate
x = sy.Symbol('x')
f = x*log(x)

# Factorial function
def factorial(n):
    if n <= 0:
        return 1
    else:
        return n*factorial(n-1)

# Taylor approximation at x0 of the function 'function'
def taylor(function,x0,n):
    i = 0
    p = 0
    while i <= n:
        p = p + (function.diff(x,i).subs(x,x0))/(factorial(i))*(x-x0)**i
        i += 1
    return p

plt.figure(figsize=(9,6))
# Plot results
def plot():
    x_lims = [0,10]
    x1 = np.linspace(x_lims[0],x_lims[1],800)
    y1 = []
    # Approximate up until 10 starting from 1 and using steps of 2
    for j in range(1,5,1):
        func = taylor(f,1,j)
        print('Taylor expansion at n='+str(j),func)
        for k in x1:
            y1.append(func.subs(x,k))
        plt.plot(x1,y1,label=str(j)+'. Ordnung')
        y1 = []
    # Plot the function to approximate (sine, in this case)
    plt.plot(x1,x1*np.log(x1),label=r'$f(x)$')
    plt.xlim(x_lims)
    plt.xlabel('x')
    plt.ylabel('y')
    plt.ylim(-5,5)
    plt.legend()
    plt.grid(True)
    plt.savefig("Aufgabe_4b.pdf")
    plt.show()

plot()
```
